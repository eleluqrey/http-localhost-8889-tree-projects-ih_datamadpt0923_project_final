```python
# Ahora tenemos un DataFrame df_comentarios_total que contiene todos los comentarios y una columna llamada
#'Sentimiento' que utilizaremos para almacenar los resultados del análisis de sentimiento.
```

```python
# Descargamos la librería que vamos a usar para el Análisis de Sentimiento
# Usaremos VaderSentiment: Bueno para textos informales, como redes sociales, y tiene una variedad de reglas y diccionarios incorporados.
```

```python
analyzer = SentimentIntensityAnalyzer()
def analizar_sentimiento_vader(texto):
    score = analyzer.polarity_scores(texto)
    if score['compound'] > 0.05:
        return 'Positivo'
    elif score['compound'] < -0.05:
        return 'Negativo'
    else:
        return 'Neutral'
```

```python
# Aplicar el análisis de sentimiento a cada comentario y almacenar los resultados en 'Sentimiento'
df_comentarios_total['Sentimiento'] = df_comentarios_total['Comentario'].apply(analizar_sentimiento_vader)
```

```python
analyzer = SentimentIntensityAnalyzer()
def analizar_sentimiento_vader(texto):
    score = analyzer.polarity_scores(texto)
    if score['compound'] > 0.05:
        return 'Positivo'
    elif score['compound'] < -0.05:
        return 'Negativo'
    else:
        return 'Neutral'
```

```python
# Aplicar el análisis de sentimiento a cada comentario y almacenar los resultados en 'Sentimiento'
df_comentarios_total['Sentimiento'] = df_comentarios_total['Comentario'].apply(analizar_sentimiento_vader)
```
